## Checking for Outliers

```python
import pandas as pd
import pandas as pd

# Assuming df is your DataFrame and you already have the lists of positive, neutral, and negative emojis
positive_emojis = ['😎', '👍', '😂', '🤣', '😘', '😏', '😜', '😊', '🙌', '💕', '😇', '🤩', '❤️', '🎉', '🤗', '🥰', '😍', '😉', '👏']
neutral_emojis = ['😳', '🙄', '🤔', '🤪', '😅']
negative_emojis = ['💔', '😭', '😢', '😡', '😔', '😅']

# Create empty DataFrames for outliers and good data
outlier_df = pd.DataFrame(columns=df.columns)
good_df = pd.DataFrame(columns=df.columns)

# Loop through the DataFrame and classify each row
for index, row in df.iterrows():
    emoji = row['Emoji']
    sentiment = row['Sentiment']

    # Check for outliers
    # Positive emojis with sentiment < 0 -> Outliers
    if emoji in positive_emojis and sentiment <= 0:
        outlier_df = pd.concat([outlier_df, pd.DataFrame([row])], ignore_index=True)

    # Neutral emojis with sentiment != 0 -> Outliers
    elif emoji in neutral_emojis and sentiment != 0:
        outlier_df = pd.concat([outlier_df, pd.DataFrame([row])], ignore_index=True)

    # Negative emojis with sentiment > 0 -> Outliers
    elif emoji in negative_emojis and sentiment >= 0:
        outlier_df = pd.concat([outlier_df, pd.DataFrame([row])], ignore_index=True)

    # If the opposite of the above conditions holds, add to good_df
    else:
        good_df = pd.concat([good_df, pd.DataFrame([row])], ignore_index=True)

print(len(df))
# Display the outliers
print("Outlier DataFrame:")
print(len(outlier_df))

# Display the good data
print("\nGood DataFrame (non-outliers):")
print(len(good_df))
```

# Splitting Training and Testing

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping

# Assuming X and y are already defined
#X_train, y_train = X_train_resampled, y_train_resampled
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12, stratify=y)
# # Convert X_train and X_test to float32 (required by TensorFlow)
X_train = np.array(X_train).astype('float32')
X_test = np.array(X_test).astype('float32')

# # Ensure y_train and y_test are of type int (if they are class labels)
y_train = np.array(y_train).astype('int')
y_test = np.array(y_test).astype('int')
```

# Building and Training a Multi-Class Classification Neural Network with Early Stopping

```python
# Build the model
model = Sequential()

# Input layer
#model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))

# Hidden layers
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))

# Output layer (assuming this is a classification problem with multiple classes)
model.add(Dense(len(np.unique(y)), activation='softmax'))  # softmax for multi-class classification

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Add early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model with early stopping
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test),
                    callbacks=[early_stopping])

# Evaluate the model on test data
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_acc}")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Evaluate the model on test data
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_acc}")

# Make predictions
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)  # Convert probabilities to class predictions

# Plot the training and validation loss and accuracy
plt.figure(figsize=(14, 6))

# Plot Loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plot Accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()

```

```python
# Print final accuracy
print(f"Final Test Accuracy: {test_acc}")

# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred_classes))
```

```python
# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_classes)

# Plot the confusion matrix using seaborn for better visualization
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
